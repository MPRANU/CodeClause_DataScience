# Import necessary libraris

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
```

## Load the dataset

```python
df = pd.read_csv('ratings.csv')
```

## Data preprocessing

```python
user_movie_ratings = df.pivot_table(index='userId', columns='movieId', values='rating')
user_movie_ratings = user_movie_ratings.fillna(0)
```

## Split the data into train and test sets

```python
train_data, test_data = train_test_split(user_movie_ratings, test_size=0.2, random_state=42)
```

## Model fitting and prediction

```python
user_similarity = cosine_similarity(train_data)
user_similarity_df = pd.DataFrame(user_similarity, index=train_data.index, columns=train_data.index)
predictions = user_similarity_df.dot(train_data).div(user_similarity_df.sum(axis=1), axis=0)
```

## Model evaluation

```python
mse = ((predictions - test_data) ** 2).mean().mean()
rmse = mse ** 0.5
print(f'RMSE: {rmse}')
```
